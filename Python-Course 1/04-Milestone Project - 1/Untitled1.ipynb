```python
## This block is written for printing instructions
def display_instruction():
    print('This is two player game where each player will have to input a number from 1 to 9 corrosponding to the positions as given below')
    print('with another number. This another number can be 1 representing * or 2 which represents x')
    print('For example a possible input can be 8,2 representing 8th position and x sign')
    horizontal = '--------------'
    vertical =   '    |    |    '
    coordinate = [2,6,10]
    vertical_modified = [' 1  |  2 |  3 ',' 4  |  5 |  6 ',' 7  |  8 |  9 ']
    iteration = 0
    for iteration1 in range(1,12):
        if iteration1%4 == 0:
            print(horizontal)
        elif iteration1 == coordinate[iteration]:
            print(vertical_modified[iteration])
            iteration += 1
            if iteration == 3:
                iteration =2
        else:
            print(vertical)
## This function is written to print the display the tic tac toe board while game is in play
display_instruction()
mark_position = []

symbols = []
final_position = [1,2,3,4,5,6,7,8,9]
final_mark = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
def display(num1,num2):
    horizontal = '--------------'
    vertical =   '    |    |    '
    coordinate = [2,6,10]
   
    global mark_position
    global symbols
    global final_position
    global final_mark
    global length
    if num1 >=10 or num1 <=0:
        print('Enter a valid number for position')
        return
    elif num1 in mark_position:
        print(' You have already entered this position please enter a new position')
        return
    elif num2 != 1 and num2 != 2:
        print(" Enter a valid number for the mark")
        return
    else:
        pass
    
    if num2 == 1:
        mark = '*'
    elif num2 ==2:
        mark = 'x'
    else:
        mark = ' '
     
    mark_position.append(num1)
    symbols.append(mark)    
    length = len(mark_position)
    index = 0
    while index <= length -1:
        num = mark_position[index]
        final_mark[num-1] = symbols[index]
        index += 1

    vertical_modified_2 = '  '+final_mark[0]+' |  '+final_mark[1]+' |  '+final_mark[2]+' '
    vertical_modified_6 = '  '+final_mark[3]+' |  '+final_mark[4]+' |  '+final_mark[5]+' '
    vertical_modified_10 = '  '+final_mark[6]+' |  '+final_mark[7]+' |  '+final_mark[8]+' '
    for t in range(1,12):
        if t%4 == 0:
            print(horizontal)
        elif t == 2:
            print(vertical_modified_2)
        elif t == 6:
            print(vertical_modified_6)
        elif t == 10:
            print(vertical_modified_10)
        else:
            print(vertical)

win_1 = 0
win_2 = 0
mark_position_1 = []
mark_position_2 = []
def check_win():
    global win_1
    global win_2
    global mark_position
    global mark_position_1
    global mark_position_2
    length = len(mark_position)
    if length <= 4:
        pass
    else:
        mark_position_1 = mark_position[0::2]
        mark_position_2 = mark_position[1::2]
        length_1 = len(mark_position_1)
        length_2 = len(mark_position_2)
    
    win_case = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],[3,6,9],[1,5,9],[3,5,7]]

    for t in range(0,8):
        
        if set(win_case[t]) <= set(mark_position_1):
            win_1 = 1
            
        elif set(win_case[t]) <= set(mark_position_2):
            win_2 = 1
        else:
            pass 
            
    if length == 9 and win_1 == 0 and win_2 == 0:
        win_1 = 2
        win_2 = 2
    
while win_1 == 0 and win_2 == 0:
    value1 = input('Input for mark position:')
    value2 = input("Input your mark symbol:")
    value1 = int(value1)
    value2= int(value2)
    display(value1,value2)
    check_win()
    
if win_1 == 1:
    print("Player 1 wins!! Hurray!!")
elif win_2 == 1:
    print('Player 2 wins!! Hurray!!')
else:
    print("Sorry but match is tie")
```
