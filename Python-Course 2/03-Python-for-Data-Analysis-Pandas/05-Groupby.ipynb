# Groupby

The groupby method allows you to group rows of data together based off a column and call aggregate functions.
<br>Aggreagate function is function that takes many values as input and gives out a single value as output. like sum, average, standard deviation, etc.


```python
import pandas as pd
import numpy as np
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
data
```

```python
type(data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

** Now you can use the .groupby() method to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object:** Output will show the memory location where object is stored.

```python
df.groupby('Company')
```

You can save this object as a new variable:

```python
by_comp = df.groupby("Company")
```

Notice that this is a groupby object.

```python
type(by_comp)
```

And then call aggregate methods off the object:

```python
by_comp.mean()
```

Person columnn is a string, so its mean has no meaning. That is why it is not showing in the output. <br>Pandas will automatically ignore non-numeric column.<br> Notice that output is a dataframe, which means that we can call dataframe methods on it.

```python
df.groupby('Company').mean()
```

```python
by_comp.sum().loc['FB']
```

```python
df.groupby('Company').sum().loc['FB']
```

More examples of aggregate methods:

```python
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```

```python
type(by_comp.describe().transpose()['GOOG'])
```
